```python
import pandas as pd
import numpy as np
bigT = pd.read_csv('./title.akas.tsv',sep='\t',low_memory=False,usecols=['titleId','title','types'])
netflix = pd.read_csv('netflix_titles.csv',usecols=['title','country'])
hulu = pd.read_csv('hulu_titles.csv',usecols=['title','country'])
disney = pd.read_csv('disney_plus_titles.csv',usecols=['title','country'])
amazon = pd.read_csv('amazon_prime_titles.csv',usecols=['title','country'])

total = set()
tit_coun = []

for i in [netflix,hulu,disney,amazon]:  
    for j in range(len(i['title'])):
        if i['title'][j] not in total:
            total.add(i['title'][j])
            tit_coun.append((i['title'][j],i['country'][j]))
            
total = pd.DataFrame(tit_coun,columns=['title','country'])
total
```

```python
import pandasql
from pandas import Series, DataFrame
f1 = DataFrame(bigT)
res = pandasql.sqldf('select distinct(total.country) from f1 join total on f1.title = total.title',globals())
res.to_csv("./4.csv")
```
